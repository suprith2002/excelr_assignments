### Q11)  Suppose we want to estimate the average weight of an adult male in    Mexico. We draw a random sample of 2,000 men from a population of 3,000,000 men and weigh them. We find that the average person in our sample weighs 200 pounds, and the standard deviation of the sample is 30 pounds. Calculate 94%,98%,96% confidence interval?

```python
from scipy import stats
import math
```

```python
#for CL 98% z score value
z=stats.norm.ppf(0.98)
```

```python
z
```

```python
data_sample=2000
```

```python
smean_x=200
```

```python
sd=30
```

```python
rv= z*(sd/math.sqrt(data_sample))
```

```python
#CI for 98%
((smean_x-rv),(smean_x+rv))
```

```python
#CI for 98%
stats.norm.interval(0.98,
                   loc=200,
                   scale=30)
```

```python
#CI for 96%
stats.norm.interval(0.96,
                   loc=200,
                   scale=30)
```

```python
#CI for 94%
stats.norm.interval(0.94,
                   loc=200,
                   scale=30)
```

2)	What can we say about the student marks? 


```python
data = np.array([34,36,36,38,38,39,39,40,40,41,41,41,41,42,42,45,49,56])
```

### Q 20) Calculate probability from the given dataset for the below cases

Data _set: Cars.csv
Calculate the probability of MPG  of Cars for the below cases.
       MPG <- Cars$MPG


```python
from scipy import stats 
cars=pd.read_csv("cars.csv")
```

```python
cars
```

#### a.	P(MPG>38)

```python
stats.norm.cdf(38,
              loc = cars.MPG.mean(),
              scale = cars.MPG.std())
```

#### b.	P(MPG<40)

```python
stats.norm.cdf(40,
              loc = cars.MPG.mean(),
              scale = cars.MPG.std())
```

#### c.    P (20<MPG<50)

```python
# p(50) - p(20)
(stats.norm.cdf(50,
              loc = cars.MPG.mean(),
              scale = cars.MPG.std()))-(stats.norm.cdf(20,
              loc = cars.MPG.mean(),
              scale = cars.MPG.std()))
```

### Q 22) Calculate the Z scores of 90% confidence interval,94% confidence interval, 60% confidence interval.

```python
print("z score of 90% :",stats.norm.ppf(0.90))
print("z score of 94% :",stats.norm.ppf(0.94))
print("z score of 60% :",stats.norm.ppf(0.60))
```

### Q 23) Calculate the t scores of 95% confidence interval, 96% confidence interval, 99% confidence interval for sample size of 25

```python
#t score for sample size of 25 whiich is degree of freedom
print("t score of 95% :",stats.t.ppf(.95,df=25-1))
print("t score of 96% :",stats.t.ppf(.96,df=25-1))
print("t score of 99% :",stats.t.ppf(.99,df=25-1))
```

### Q 24)   A Government company claims that an average light bulb lasts 270 days. A researcher randomly selects 18 bulbs for testing. The sampled bulbs last an average of 260 days, with a standard deviation of 90 days. If the CEO's claim were true, what is the probability that 18 randomly selected bulbs would have an average life of no more than 260 days

```python
#t-score for x=260 // formula : t-score= (smean-pmean)/(S_std/sqrt(n))
```

```python
t_score = (260-270)/(90/math.sqrt(18))
```

```python
t_score
```

```python
#to find probability P(x)<=260 ?
```

```python
stats.t.cdf(t_score,df=18-1)  #for randomly selected 18 bulbs
```
